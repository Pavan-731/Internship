Q.2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
and make data frame.

```python
page2 = requests.get ('https://www.imdb.com/list/ls091520106/')
page2

soup2 = BeautifulSoup(page2.content)
    


# Top 100 Movies Name
name=soup2.find_all("h3", class_="lister-item-header")
Top_100_film_name = [] 
for i in name:
    for j in i.find_all("a"):
        Top_100_film_name.append(j.text)


# Year of Release 
year= soup2.find_all("span",class_="lister-item-year text-muted unbold")
Year_of_Release = [] 
for i in year:
    a=i.text.replace('(','')
    Year_of_Release.append(a.replace(')',''))    
    
# IMDB Rating 
rating= soup2.find_all("div",class_="ipl-rating-star small")
IMDB_Rating = [] 
for i in rating:
    IMDB_Rating.append(float(i.text))

    
#Making DataFrame
IMDB_top_100= pd.DataFrame({})
IMDB_top_100['Movies_Name'] =Top_100_film_name
IMDB_top_100['Year_of_Release'] =Year_of_Release
IMDB_top_100['IMDB_Rating'] =IMDB_Rating
IMDB_top_100
    
    
    
```

Q.7 Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
and the code for the video from the link for the youtube video from the post.

```python
page11 = requests.get ('https://coreyms.com/')

soup11 = BeautifulSoup(page11.content)


articles = soup11.find_all('article')

data = []
for article in articles:
    header = article.find("header")
    
    # Title
    title = header.find("h2").text
    # Date
    date = header.find("time", class_="entry-time").text
    div_block = article.find("div", class_="entry-content")
    # Paragraph
    paragraph_content = div_block.find("p").text
    try:
        youtube_link = div_block.find("iframe", class_="youtube-player")['src']
        video_code = youtube_link.split("?")[0].split("/")[-1]
    except:
        video_code = "NA"
    # Video Code
    data.append([title,date, paragraph_content, video_code])
    # printing the dataframe. 
columns = ["Title","Date", "Content","Video Code"]
df = pd.DataFrame(data, columns=columns)
df
```
